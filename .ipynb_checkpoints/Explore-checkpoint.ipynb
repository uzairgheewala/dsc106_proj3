```python
def plot_plants_over_gibs(
    date="2010-01-01",
    layer_name="BlueMarble_NextGeneration",
    pop_col="p10_30"  # 2010 population within 30 km
):
    # Connect to GIBS WMS
    wms = WebMapService(
        "https://gibs.earthdata.nasa.gov/wms/epsg4326/best/wms.cgi",
        version="1.3.0"
    )

    bbox = (-180, -90, 180, 90)
    width, height = 2048, 1024

    img = wms.getmap(
        layers=[layer_name],
        styles=["default"],
        srs="EPSG:4326",
        bbox=bbox,
        size=(width, height),
        format="image/jpeg",
        transparent=True,
        time=date,
    )
    img_arr = np.array(Image.open(BytesIO(img.read())))

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.imshow(img_arr, extent=bbox, origin="upper")

    # scale population so sizes are reasonable
    pop_millions = plants_df[pop_col] / 1e6
    sizes = 10 + pop_millions * 4  # tweak multiplier as needed

    sc = ax.scatter(
        plants_df["Longitude"],
        plants_df["Latitude"],
        s=sizes,
        c=plants_df["NumReactor"],
        cmap="viridis",
        alpha=0.85,
        edgecolor="black",
        linewidth=0.2,
    )

    ax.set_xlim(-180, 180)
    ax.set_ylim(-90, 90)
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    ax.set_title(
        f"Nuclear Power Plants â€“ size = pop within 30 km ({pop_col}), "
        f"color = # reactors â€“ {date}"
    )

    cbar = plt.colorbar(sc, ax=ax, label="Number of reactors")
    plt.tight_layout()
    plt.show()

plot_plants_over_gibs()
```

```python
def plot_exposure_hotspots(
    date="2010-01-01",
    layer_name="BlueMarble_NextGeneration",
    pop_col="p10_30"  # 2010 population within 30 km
):
    # ----- background from GIBS -----
    wms = WebMapService(
        "https://gibs.earthdata.nasa.gov/wms/epsg4326/best/wms.cgi",
        version="1.3.0"
    )

    bbox = (-180, -90, 180, 90)
    width, height = 2048, 1024

    img = wms.getmap(
        layers=[layer_name],
        styles=["default"],
        srs="EPSG:4326",
        bbox=bbox,
        size=(width, height),
        format="image/jpeg",
        transparent=True,
        time=date,
    )
    img_arr = np.array(Image.open(BytesIO(img.read())))

    fig = plt.figure(figsize=(13, 6))
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.set_global()

    # satellite image
    ax.imshow(
        img_arr,
        extent=bbox,
        origin="upper",
        transform=ccrs.PlateCarree(),
        zorder=0,
    )

    # dark coastlines / borders so theyâ€™re visible
    ax.coastlines(resolution="110m", linewidth=0.6, color="black", zorder=2)
    ax.add_feature(cfeature.BORDERS, linewidth=0.4, edgecolor="black", zorder=2)

    # ----- exposure metric & categories -----
    df = plants_df.copy()
    df["pop30_m"] = df[pop_col] / 1e6
    q50 = df["pop30_m"].quantile(0.5)
    q90 = df["pop30_m"].quantile(0.9)

    df["exposure_cat"] = pd.cut(
        df["pop30_m"],
        bins=[0, q50, q90, df["pop30_m"].max()],
        labels=["Low", "Medium", "High"],
    )

    colors = {"Low": "#fee8c8", "Medium": "#fdbb84", "High": "#e34a33"}
    df["color"] = df["exposure_cat"].map(colors)

    sizes = 15 + df["pop30_m"] * 5

    sc = ax.scatter(
        df["Longitude"],
        df["Latitude"],
        s=sizes,
        c=df["color"],
        alpha=0.9,
        edgecolor="black",
        linewidth=0.25,
        transform=ccrs.PlateCarree(),
        zorder=3,
    )

    # highlight top 10 exposure plants
    top = df.nlargest(10, "pop30_m")
    for _, row in top.iterrows():
        ax.text(
            row["Longitude"],
            row["Latitude"] + 2,
            row["Plant"],
            fontsize=7,
            ha="center",
            va="bottom",
            transform=ccrs.PlateCarree(),
            zorder=4,
        )

    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")

    title = "Global hotspots of population exposure to nuclear power plants (30 km, 2010)"
    ax.set_title(title, fontsize=13, pad=12)

    # legend for exposure categories
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], marker="o", color="w",
               label="Low (â‰¤ median)", markerfacecolor=colors["Low"],
               markeredgecolor="black", markersize=8),
        Line2D([0], [0], marker="o", color="w",
               label="Medium (50â€“90th pct)", markerfacecolor=colors["Medium"],
               markeredgecolor="black", markersize=8),
        Line2D([0], [0], marker="o", color="w",
               label="High (top 10%)", markerfacecolor=colors["High"],
               markeredgecolor="black", markersize=8),
    ]
    ax.legend(handles=legend_elements, loc="lower left", fontsize=8)

    # fig.text(
    #     0.5, -0.02,
    #     "Are nuclear exposure hotspots concentrated in only a few regions, or globally distributed?",
    #     ha="center", va="top", fontsize=10,
    # )

    plt.tight_layout()
    plt.show()

plot_exposure_hotspots()
```
